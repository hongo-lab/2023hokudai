### 可変長引数

````python
def func_args(*args): # 仮引数(args)の先頭に「アスタリスク(*)」を付けると、数を決めずに引数を渡せる
  print('args:', args)
  print('args[1]:', args[1])

func_args(1, 2, 3, 4, 5) # 引数はリストとして保存され、関数内で参照される
````

````python
def func_kwdargs(**kwdargs): # ２つのアスタリスク(**)を付けると、数を決めずに「キーワード引数」を渡せる
  print('kwdargs:', kwdargs)
  print('kwdargs[a]:', kwdargs['a'])

func_kwdargs(a=1, b=2, c=3, d=4, e=5) # 引数は辞書型(キーと値の組)として保存され、関数内で参照される
````

### ファイル読み込み

````python
path = './Cu.pz-dn-rrkjus_psl.0.2.UPF'
with open(path) as f: # ファイルオープン
    line = f.readline() # 1行読み込み
    print(line)
````

````python
with open(path) as f:
    for line in f: # 1行ずつ読み込み
      print(line)
````

````python
with open(path) as f:
  for line in f:
    line = line.rstrip() # 開業コードを削除するために、rstrip()メソッドを利用
    print(line)
````

````python
with open(path) as f:
  for line in f:
    line = line.rstrip()

    if line.startswith('<'):
      print(line)
````
